# SQL Subqueries, Common Table Expressions (CTE), and Temp Tables

Sometimes SQL queries need to go beyond just joining existing tables together. We need to manipulate existing tables to be able to generate the data needed. There are three ways to accomplish that:

1. Subqueries.

2. Common Table Expressions (CTE).

3. Temporary Tables (Temp Tables).

All of these approaches allow you to structure your queries in a way that breaks down complex operations into manageable steps, but they do so in slightly different ways.

Temp tables are more advanced and are outside the scope of the class. See the supplemental notebook if you are interested in learning more about them.

## Subqueries

A subquery is a query nested within another SQL query, typically within the SELECT, WHERE, or FROM clause.

Subqueries allow you to perform calculations or filtering within the context of the outer query.

They are particularly useful when you need to filter results based on aggregated values like maximum, minimum, and average or are pivoting the data from rows to columns like if a customer's telephone numbers are stored in a table, and you want to have their home / work / mobile numbers in their own columns.

Subqueries are very useful, but tying them to individual rows in your SELECT statement is a common performance trap. It forces SQL to recalculate the math for every single row. Instead, use subqueries to calculate your totals upfront and then join the results!

Visit the [w3resource's SQLite Subqueries page](https://www.w3resource.com/sqlite/sqlite-subqueries.php) to learn more.



## Common Table Expressions (CTEs)

A Common Table Expression (CTE), introduced by the WITH clause, is like a subquery, but one that you can reference multiple times in your main query.

CTEs can improve the readability and maintainability of SQL code, especially for complex queries, by allowing you to break down operations into logical steps.

They are helpful when the same calculation needs to be referenced multiple times within a query, or when you need to calculate aggregate data but return the information side-by-side with individual rows.

Visit the [SQLite Documentation's CTE page](https://www.sqlite.org/lang_with.html) to learn more.

### Subqueries

- Are created/defined and executed in the same statement.

- Are defined `inline`, within your main query, and as part of that main query.

- Are only executed where defined, not reusable.

### CTEs

- Are created/defined at the top of the SQL statement using the WITH clause.

- Act as a temporary, named result set that exists only for the duration of that specific query.

- Are syntactically reusable, meaning you can write the logic once at the top, and reference its name multiple times in your main query.

- Practically, this means CTEs keep your code DRY (Don't Repeat Yourself). Instead of copying and pasting the same subquery three times, you define it once as a CTE and simply JOIN to it wherever you need it.
