### Let's look at the testing paradigm in detail.

The below describes the 4 data elements that will be used to test your function.

This cell appears in every exercise on the exams, and in most exercises in the homework notebooks.

****

The test cell below will check your solution against several randomly generated test cases. If your solution does not pass the test (or if you're just curious), you can look at the variables used in the latest test run. They are automatically imported for you as part of the test.

- `input_vars` - Dictionary containing all of the inputs to your function. Keys are the parameter names.
- `original_input_vars` - Dictionary containing a copy of all the inputs to your function. This is useful for debugging failures related to your solution modifying the input. Keys are the parameter names.
- `returned_output_vars` - Dictionary containing the outputs your function generated. If there are multiple outputs, the keys will match the names mentioned in the exercrise instructions.
- `true_output_vars` - Dictionary containing the outputs your function **should have** generated. If there are multiple outputs, the keys will match the names mentioned in the exercrise instructions.

All of the test cells in every exam will use this same format, and you can expect a similar format on your homework notebooks as well.

#### These are named the `TEST CASE VARIABLES`, and note that they are all dictionaries. You will use dictionary functions, which we will be covering in detail next week, in your data troubleshooting.

****

### Recall from above that the test call has two test portions that are performed on your function.

#### The first portion of the test is to ensure that your function does not modify any of the input variables.

1. The test cell compares the `original_input_vars` to the `input_vars`. If they are the same, then this portion of the test passes. If they are different in any way, then this portion of the test fails.


2. If this portion of the test fails, it means that your function has modified the inputs in some way. ***We always advise students to make a copy of the input variables within their functions, and work only with those copies.***


3. If this portion of the test fails, the test will not continue on to the next step in the testing process.


4. What you want to do is go up to your code and find the line that is doing the modification.

#### The second portion of the test is to ensure that your function returns the correct solution.

1. The test cell compares the `returned_output_vars` to the `true_output_vars`. If they are the same, then this portion of the test passes. If they are different in any way, then this portion of the test fails.


2. Same as above, in Module 0 of the course, when the outputs of your solution are being tested, the test is looking at 3 things:

    1. The data types between the two solution outputs are the same.
    2. If the solution is longer than a single value (list/dict/set), the `true_outpu_vars` and `returned_output_vars` variables are the same length.
    3. The actual data itself is the same, no matter the data structure (list, dict, set, etc)


3. This portion of the test will return one of two results:

    1. A `passed` message. You have passed the exercise and can move on to the next exercise.
    2. A message stating that the `returned_output_vars` does not match the `true_output_vars`. If this is the case, then you will need to perfom data troubleshooting and modify your function.

### There are two general methods to troubleshoot, when your function fails for one of the above test portions.

1. You can visually inspect and compare the `test case variables`. This approach has some value, and it is probably a good starting point to compare for your troubleshooting.


#### However, you `SHOULD NOT` rely soley on a visual inspection for your data troubleshooting, because a visual inspection of the data does not test all 3 scenarios above. Additionally, your visual inspection may not `see` all of the differences.


2. You can run some code to test the 3 testing scenarios above. Later in this session, we will be providing template code to do this. You can copy and paste this template code into your exam and run it to test your outputs.
