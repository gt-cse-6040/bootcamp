### Finding our Favorite Movies: Solution

Here's one possible solution to the problem above.

```python
def favorite_movies_solution(log, score):
    fav_movies = []
    for movie in log:
        if log[movie] == score:
            fav_movies.append(movie)
    return fav_movies

favorite_movies_solution(my_movies, 5)
```
